Object Oriented Programming is the process of organizing code around data - or 'objects'

It binds data (attributes of an object) and behavior (methods of an object) into cohesive, reusable units aka objects so we can say an object is what is has and what it does.

An ATTRIBUTE is a fancy word for VARIABLE attached to an OBJECT
A METHOD is function that an OBJECT can do or the behavior of the OBJECT
We can also say that an object is a combination of data/variable tied to a functionality
