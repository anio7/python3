to create or wipe a dictionary we use curly brackets-
dic_Name = {} #empty dictionary

```python
first_dic = {}
print(first_dic)
```

to add items to a dictionary we do something similar and whatever is on the RHS of the "=" is added to the dictionary
dic_name[key_name] = ""

```python
first_dic["live"] = "live today!"
first_dic["love"] = "love unconditionally!"
first_dic["learn"]= "learn quickly!"
first_dic["laugh"]= "laugh loudly!"

print(first_dic)
```

to retrieve items from a dictionary you use the key in square brackets
dic_name[key_name]

```python
first_dic["love"]
```


for iteration through a dictionary there is a special way to get the key or value  or both, using

for i in dic_name:
    do something
    to get the keys printed print(i)
    to get the values printed use print(dic_name[i])
    to get both key & values printed use print(i,dic_name[i])

```python
student_scores = {
    'Harry': 88,
    'Ron': 78,
    'Hermione': 95,
    'Draco': 75,
    'Neville': 60
}
for k in student_scores:
    print(k,student_scores[k])
```
