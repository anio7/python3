A list comprehension is creating a list from a previous list.
eg 

{var_name} = [{expression},{for loop statement}]
whatever is in the expression is appended to the list eg. n * 2 or n
basic for loop statement needed 

new_list = [new_item for item in list]

```python
number = [1,2,3]
new_list = []
for n in numbers:
    add_1 = n + 1
new_list.append(add_1)

#[{expression},[for loop statement]]
numbers = [1,2,3]
new_list = [n + 1 for n in numbers]
```

```python
#you can also you for...range() as well...
range_list = [n for n in range(1,6)]
print(range_list)
```

There is also conditional list comprehension

var_name = [{expression}{for loop}{condition}]
    whatever is in the expression is appended to the list eg. n * 2 or n
    the for loop can also be for ...range()
    condition is whatever needs to be evaluated.
new_list = [new_item | for item in list | if test]

```python
names = ["alex", "beth", "caroline", "dave", "elanor", "freddie"]
short_names = [name for name in names if len(name)< 5]
print(short_names)
```
