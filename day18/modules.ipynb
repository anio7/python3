For importing modules - 
variable = ModuleName.ClassName() - imports a class from a module if you are going to use it once.
from module import ClassName - imports a class if you are going to use it multiple times
from module import * - imports everything

ALIAS modules 
import module as <alias>

some modules you have to INSTALL before importing
pip install <module>
