A CONSTRUCTOR is a part of the blueprint which specifies what should happen when the object is being initialized. We do this by using a special function called init aka initialize.
When a object is being initialized we can set variables or counters to their starting values.

Self is the object being created or initialized, you can add more parameters. Once you received the data you can use it to set it the objects Attributes and initialize them.
A Method always needs a self parameter as its first parameter

```python
class User:

#special method
    def __init__(self, user_id, username):
        #initialize attributes
        print("welcome to classes")
        self.id = user_id
        self.username = username.capitalize()
        self.followers = 0
        self.following = 0
        
    #initialize the method
    def follow(self,user):
        user.followers += 1
        self.following += 1

user_1 = User("001","anderson")
user_2 = User("002","niomi")
user_1.follow(user_2)


print(user_1.id,user_1.username)
print(user_1.following)
print(user_2.following)
```
