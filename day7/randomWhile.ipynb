A simple example of a while loop with the condition before.

```python
gameOver = False
while not gameOver:
    print("Code goes here")
    playAgain = input("Do you want to play again?")
    #typing 'y' here will keep you in the while loop
    if playAgain =="n":
        print("Its over now")
        gameOver = True
print("Good bye")
```
