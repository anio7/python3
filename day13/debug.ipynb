using the try - except is used for catching bugs.

```python
try:
    age = int(input("How old are you? "))
except ValueError:
    print("you have typed in a invalid number"
          "Please try again with a numerical number"
          "please enter another number"
          )
    age = int(input("how old are you? "))
    
if age > 18:
    print(f"You can drive at age {age}")
```

use print() as a form for debugging since it shows you, your variables before you insert them into the code.

```python
words_per_page = 0
pages = int(input("what is the number of pages in the book? "))
words_per_page = int(input("Number of words per page: "))
total_words = pages * words_per_page
print(words_per_page)
print(pages)
print(f"the total amount of words are {total_words}")
```
