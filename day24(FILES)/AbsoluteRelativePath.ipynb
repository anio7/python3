Absolute File Path always start from the root or origin
it denotes the / as root.

eg /work/project/talk.ppt

Relative File Path starts from the current working directory. the ./ states to look into the working directory 
./talk.ppt
to go one step up we write
.. /report.doc

in mac is forward slash(/) and windows is backward slash (\) but in python its not a problem.

To go up 2 levels we use ../ ../ for 2 levels.
