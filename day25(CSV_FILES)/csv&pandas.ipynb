Comma Separated Values (CSV)

```python
import csv

with open("./weather_data.csv") as w_data:
    data = csv.reader(w_data)
    temps = []
    for row in data:
        print(row)
        if row[1] != "temp":
            temps.append(int(row[1]))
    print(temps)
```
