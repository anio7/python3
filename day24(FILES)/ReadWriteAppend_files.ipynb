Reading and Writing to a file

To open and read a file that already been created in python we can use open(<file_name>) and save it to a variable.

To close a file in python we use close() to release the resources of the computer.



```python
file = open("my_text.txt")
contents = file.read()
print(contents)
file.close()
```

for a better way for opening and closing a file automatically we use 'with' .... as .... keywords

```python
with open("my_text.txt") as file:
    contents = file.read()
    print(contents)
```

To write to a file, you need to change the mode of the file from read only -'r' to write -'w'. This overwrites all the previous text inside the file. If the file doesn't exist it will create a file.

```python
with open("my_text.txt",mode = "w") as file:
    file.write("new text")

#to view the contents you must open the file again.
file = open("my_text.txt")
contents = file.read()
print(contents)
```

If you do not want to delete the contents you have to append to the file by which you change the mode to "a", this adds it to the end of the contents.

```python
with open("my_text.txt",mode = "a") as file:
    contents = file.write("new text")
    print(contents)
```

Using .readlines() causes the data to be displayed as a list.
