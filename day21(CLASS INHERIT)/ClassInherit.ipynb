There is something called class inheritance and in order for a class to inherit from another class we have to call the class within the new class parentheses and place super(). in front of the __init__

```python
class ClassName(Other_Class_Name):  #insert the class you would like to inherit from into the parentheses
    def __init__(self):
        super().__init__() #this gives us access to the ATTRIBUTES & METHODS of the INHERITED CLASS. Super refers to super class.
```

```python
class Animal:
    def __init__(self):
        self.num_eyes = 2

    def breathe(self):
        print("inhale...exhale")

class Fish(Animal):
    def __init__(self):
        super().__init__()

    def breathe(self):
        super().breathe() #this prints the initial breathe method and updated print("doing this underwater")
        print("doing this underwater\n")

    def swim(self):
        print("moving in water \n")

animal = Animal()
fish = Fish()

animal.breathe()
fish.swim()
fish.breathe()
```
