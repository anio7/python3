```python
names = ["alex", "beth", "caroline", "dave", "elanor", "freddie"]
import random

student_scores = {student:random.randint(1,100) for student in names}
print(student_scores)

passed_students = {student:score for (student,score) in student_scores.items() if score >= 80}
print(passed_students)
```

```python
sentence = "What is the Airspeed Velocity of an Unladen Swallow?"
sentence = sentence.split()
result = {word:len(word) for word in sentence}
print(result)


# new_dict = {new_key:new_value for item in list}
# new_dict = {new_key:new_value for (key, value) in dict.items()}
# new_dict = {new_key:new_value for (key, value) in dict.items() if test}
```

```python
weather_c = {"Monday": 12, "Tuesday": 14, "Wednesday": 15, "Thursday": 14, "Friday": 21, "Saturday": 22, "Sunday": 24}
day = {day for day in weather_c.keys()}
temp_c = {day for day in weather_c.values()}
print(temp_c)
temp_f = {(n * 9/5 + 32) for n in weather_c.values()}
print(temp_f)
weather_f = {day:(temp * 9/5 + 32) for (day, temp) in weather_c.items()}


print(weather_f)
```
